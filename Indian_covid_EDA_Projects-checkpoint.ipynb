```python
total_death=df.Deaths.sum()
total_cure=df.Cured.sum()
total_confirm=df.Confirmed.sum()
```

```python
total_death,total_cure,total_confirm
print('Total Deaths are = {}'.format(total_death))
print('Total Cure are = {}'.format(total_cure))
print('Total Confirmed are = {}'.format(total_confirm))
```

```python
error = total_confirm - (total_death+total_cure)
print('So, there seems to have some missing data around {}, regarding the people who where positive.'.format(error))
```

* Error is the data which is not known

```python
death_rate=(total_death*100)/total_confirm
death_rate
print("The overall death rate reported in INDIA is = {:.2f}%".format(death_rate))
```
